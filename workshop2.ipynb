1. Write a Python function that takes a list and returns a new list with distinct elements from the first list. Sample List : [1,2,3,3,3,3,4,5] Unique List : [1, 2, 3, 4, 5]

```python
def unique_list(lst):
    new_list = []
    for i in lst:
        if i not in new_list:
            new_list.append(i)
    return new_list

# Example
sample = [1,2,3,3,3,3,4,5]
print(unique_list(sample))

```

2. Write a Python function that takes a number as a parameter and checks whether the number is prime or not.

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True


print(is_prime(7))
```

3. Write a program to print twin primes less than 1000. If two consecutive odd numbers  are both prime then they are known as twin primes

```python
def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

for i in range(3, 1000, 2):
    if is_prime(i) and is_prime(i+2):
        print(i, i+2)
```

4. A positive integer is called a perfect number if it is equal to the sum of all of its divisors, including 1 but excluding the number itself. For example, 6 = 1 + 2 + 3. Write a function sumOfDivisors that takes a positive integer n and returns the sum of all its divisors (excluding n).

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

# Example
print(sumOfDivisors(6))

```

```python
def sumOfDivisors(n):
    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i
    return total

num = 7
while True:
    if sumOfDivisors(num) == num:
        print("Next perfect number:", num)
        break
    num += 1
```

6. Write the python program to find the mean median from the following list: Marks = [20, 30, 35, 55, 70, 80, 90] Do not use the library function if there is one.

```python
marks = [20, 30, 35, 55, 70, 80, 90]

# Mean
total = 0
for i in marks:
    total += i
mean = total / len(marks)

# Median
marks.sort()
n = len(marks)

if n % 2 == 0:
    median = (marks[n//2] + marks[n//2 - 1]) / 2
else:
    median = marks[n//2]

print("Mean:", mean)
print("Median:", median)
```

7. Write a function that converts a decimal number to binary number.


```python
def decimal_to_binary(n):
    binary = ""
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2
    return binary

# Example
print(decimal_to_binary(10))
```

8. Write a function prodDigits() that inputs a number and returns the product of digits of that
number.


```python
def prodDigits(n):
    product = 1
    while n > 0:
        digit = n % 10
        product *= digit
        n = n // 10
    return product

# Example
print(prodDigits(123))
```

9. Write a function that will take in an unknown number of arguments and multiply all of them
together and run the function for these 2 sets of numbers:
a. 1,2,3,4,5
b. 12,13,14

```python
def multiply_all(*args):
    result = 1
    for num in args:
        result = result * num
    return result

# a. 1,2,3,4,5
print("Result a:", multiply_all(1,2,3,4,5))

# b. 12,13,14
print("Result b:", multiply_all(12,13,14))
```
