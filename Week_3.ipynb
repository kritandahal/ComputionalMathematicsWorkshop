```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 20)

y1 = 5*x**3 + 2*x - 1
y2 = -2*x**3 + x**2 + 100
y3 = 2*np.pi*x + 20

plt.plot(x, y1, label="5x^3 + 2x -1")
plt.plot(x, y2, label="-2x^3 + x^2 +100")
plt.plot(x, y3, label="2πx + 20")

plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-5, 5, 20)

plt.subplot(3,1,1)
plt.plot(x, 5*x**3 + 2*x - 1)
plt.title("5x^3 + 2x -1")

plt.subplot(3,1,2)
plt.plot(x, -2*x**3 + x**2 + 100)
plt.title("-2x^3 + x^2 +100")

plt.subplot(3,1,3)
plt.plot(x, 2*np.pi*x + 20)
plt.title("2πx + 20")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3*np.pi, 3*np.pi, 100)
y = np.cos(x)

plt.plot(x, y)
plt.title("y = cos(x)")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

a, b = 2, 2
r = 4

x = []
y = []

t = 0
while t <= 2*math.pi:
    x.append(a + r*math.cos(t))
    y.append(b + r*math.sin(t))
    t += 0.1

plt.plot(x, y)
plt.gca().set_aspect('equal')
plt.title("Circle")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.linspace(0, 2*np.pi, 100)

# Circle
x1 = 2 + 4*np.cos(t)
y1 = 2 + 4*np.sin(t)

# Ellipse
x2 = 2 + 6*np.cos(t)
y2 = 2 + 3*np.sin(t)

plt.plot(x1, y1, label="Circle")
plt.plot(x2, y2, label="Ellipse")

plt.gca().set_aspect('equal')
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

p, q = 1, 2
a = 0.5

x = np.linspace(-10, 10, 100)
y = p + 4*a*(x - q)**2

plt.plot(x, y)
plt.title("Parabola")
plt.grid()
plt.show()
```
