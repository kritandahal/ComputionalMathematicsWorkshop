```python
 import matplotlib.pyplot as plt
import random

# Generate 10 random values
x = [random.randint(0,50) for i in range(10)]
y = [random.randint(0,50) for i in range(10)]

# Colors list
colors = ["red","green","blue","yellow","pink",
          "orange","purple","brown","cyan","magenta"]

# Scatter plot with colors
plt.scatter(x, y, c=colors)

# Labels and title
plt.xlabel("Time taken in seconds")
plt.ylabel("Distance reached")
plt.title("Buzzer Game Results")

plt.show()
```

```python
import matplotlib.pyplot as plt

animals = ['Lion', 'Tiger', 'Elephant', 'Deer']

pop_2017 = [20, 30, 25, 25]
pop_2018 = [25, 25, 30, 20]
pop_2019 = [30, 20, 25, 25]
pop_2020 = [35, 15, 30, 20]
pop_2021 = [40, 20, 20, 20]
pop_2022 = [45, 15, 25, 15]

years = [pop_2017, pop_2018, pop_2019, pop_2020, pop_2021, pop_2022]
year_labels = ['2017', '2018', '2019', '2020', '2021', '2022']

fig, axes = plt.subplots(2, 3, figsize=(12, 8))

for i, ax in enumerate(axes.flat):
    ax.pie(years[i], labels=animals, autopct='%1.1f%%')
    ax.set_title(year_labels[i])

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
cars = ['Red', 'Silver', 'Blue', 'White', 'Green']
values = [123, 230, 78, 193, 12]

# ---------------- a) Bar Chart ----------------
plt.figure()
plt.bar(cars, values)
plt.title("Car Color Count (Bar Chart)")
plt.xlabel("Car Colors")
plt.ylabel("Number Seen")
plt.show()

# ---------------- b) Width Effect ----------------
plt.figure()
plt.bar(cars, values, width=0.3)
plt.title("Bar Chart with width = 0.3")
plt.show()

# Effect: Bars become thinner (narrow)

# ---------------- c) Horizontal Bar ----------------
plt.figure()
plt.barh(cars, values)
plt.title("Horizontal Bar Chart")
plt.show()

# Effect: Bars are horizontal instead of vertical

# ---------------- d) Color Matching ----------------
colors = ['red', 'silver', 'blue', 'white', 'green']

plt.figure()
plt.bar(cars, values, color=colors)
plt.title("Bar Chart with Matching Colors")
plt.show()

# ---------------- e) Pie Chart ----------------
plt.figure()
plt.pie(values, labels=cars, autopct='%1.1f%%')
plt.title("Car Color Distribution")
plt.show()

# ---------------- f) Pie Chart Variations ----------------

# A - Explode
plt.figure()
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(values, labels=cars, explode=myexplode, autopct='%1.1f%%')
plt.title("Exploded Pie Chart")
plt.show()

# B - Explode + Shadow
plt.figure()
myexplode = [0.3, 0, 0, 0, 0]
plt.pie(values, labels=cars, explode=myexplode, shadow=True, autopct='%1.1f%%')
plt.title("Exploded Pie with Shadow")
plt.show()

# C - Legend
plt.figure()
plt.pie(values, labels=cars, autopct='%1.1f%%')
plt.legend(title="Car Colours:")
plt.title("Pie Chart with Legend")
plt.show()

# D - Custom Colors
plt.figure()
mycolours = ["magenta", "green", "cyan", "red", "yellow"]
plt.pie(values, labels=cars, colors=mycolours, autopct='%1.1f%%')
plt.title("Pie Chart with Custom Colors")
plt.show()
```
