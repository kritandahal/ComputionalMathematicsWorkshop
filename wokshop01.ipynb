```python
def myPow(x, n):
    result = 1
    i = 0

    while i < n:
        result = result * x
        i += 1

    return result

x = int(input("Enter base: "))
n = int(input("Enter exponent: "))

print("Result:", myPow(x, n))
```

```python
def fib(n):
    a = 0
    b = 1
    count = 0

    while count < n:
        print(a, end=" ")
        c = a + b
        a = b
        b = c
        count += 1

n = int(input("Enter number of terms: "))
fib(n)
```

```python
def SumDigits(n):
    s = 0

    while n > 0:
        digit = n % 10
        s = s + digit
        n = n // 10

    print("Sum of digits:", s)

num = int(input("Enter number: "))
SumDigits(num)
```

```python
def squaredSum(arr):
    total = 0

    for num in arr:
        total = total + num * num

    return total

numbers = [2, 3, 4]

print("Sum of squares:", squaredSum(numbers))
```

```python
def sumDigits(num):
    total = 0

    while num > 0:
        digit = num % 10
        total = total + digit
        num = num // 10

    return total

n = int(input("Enter number: "))
print("Sum of digits:", sumDigits(n))
```
